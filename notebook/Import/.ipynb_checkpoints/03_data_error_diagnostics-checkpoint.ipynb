```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('max_column', 100)
```

```python
import_path = '/Users/zihaoxu/R_repos/sgv_bike/import/'
dataset_path = '/Users/zihaoxu/R_repos/sgv_bike/datasets/'
```

```python
# Read in the data
bike = pd.read_csv(import_path + 'bike.csv')
bike.head()

ped = pd.read_csv(import_path + 'ped.csv')
ped.head()
```

```python
print(list(bike))
```

```python
# Sum up the four directional vars
bike['total_counts'] = [a+b+c+d for a,b,c,d in zip(bike['NS'],bike['SN'],bike['EW'],bike['WE'])]
# Create list of potential problems
potential_problem_vars = ['Female', 'Sidewalk', 'Wrongway', 'Other_Var1_Count', 'Other_Var2_Count']
issue_nums = []

# Count number of issues per observation
for i in range(len(bike)):
    issue_num = 0
    for var in potential_problem_vars:
        if bike.loc[i, var] > bike.loc[i, 'total_counts']:
            issue_num += 1
    issue_nums.append(issue_num)   
    
bike['issue_nums'] = issue_nums
problem = bike[bike['issue_nums'] > 0][['IntervalID', 'location_ID'] + potential_problem_vars + ['NS', 'SN', 'EW', 'WE', 'total_counts', 'issue_nums']]
problem
```

```python
print("Number of issues with bike dataset: ", len(problem))
problem.to_csv(dataset_path + 'raw/bike_problems.csv', index = None)
```
