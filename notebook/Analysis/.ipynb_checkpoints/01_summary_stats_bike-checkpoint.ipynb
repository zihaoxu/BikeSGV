```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('max_column', 100)
```

```python
mst_path = '/Users/zihaoxu/R_repos/sgv_bike/datasets/mst/'
fig_path = '/Users/zihaoxu/R_repos/sgv_bike/export/01_summary_stats_bike/'
```

```python
df = pd.read_csv(mst_path + 'clean_bike.csv')
df.head()
```

```python
num_cols = ['female', 'sidewalk', 'wrongway', 'no_helmet', 'child', 'total_count']
```

```python
res = df.groupby('city').sum()[num_cols]
res = res.reset_index()
for col in num_cols[:-1]:
    res[col+'_perc'] = res[col] / res['total_count']
res
```

```python
temp = res.sort_values('total_count', ascending = False)[['city', 'total_count']].head(10)
temp.to_csv(fig_path + 'total_riders.csv', index = None)

sns.barplot(data = temp, y = 'city', x = 'total_count', orient = "h")
sns.despine()
plt.ylabel('City Name')
plt.xlabel('Total Number of Riders')
plt.title('Total Number of Riders by City (Top 10)')
plt.tight_layout()
plt.savefig(fig_path + 'total_riders', fmt = 'png', dpi = 300)
plt.show()
```

```python
def display_top(df, col, top_num = 5):
    ''' Get the top 5 cities in terms of col_prec or proportion for the given col '''
    return df.sort_values(col+'_perc', ascending = False)[['city', col+'_perc', col, 'total_count']].head(top_num)

def plot_proportion(col):
    ''' Plots the proportion of col for the summary stats '''
    temp = display_top(res, col)
    temp.to_csv(fig_path + col + '_city.csv', index = None)
    sns.barplot(data = temp, y = 'city', x = col+'_perc', orient = "h")
    sns.despine()
    plt.ylabel('City Name')
    plt.xlabel(col.title() + ' Proportion')
    plt.title('Percentage of ' + col.title() + ' Riding by City (Top 5)')
    plt.tight_layout()
    plt.savefig(fig_path + col + '_city', fmt = 'png', dpi = 300)
    plt.show()
```

```python
for col in num_cols[:-1]:
    plot_proportion(col)
```

# Look at bikewaytype_options

```python
df['bikewaytype_options'].value_counts()
```

```python
# Create new column and replace column entries
df['has_sidewalk_riding'] = np.where(df['sidewalk'] > 0, "Yes", "No")
df['bikewaytype_options'] = np.where(df['bikewaytype_options'] == "bikelane_paintedbuffer;", 
                                     "Bikelane with Painted Buffer",
                                     np.where(df['bikewaytype_options'] == "bikeroute_sharrows;",
                                              "Bikeroute with Sharrows", "None"))
```

```python
tab = pd.crosstab(df['bikewaytype_options'], df['has_sidewalk_riding'], normalize = 'index')
# plt.figure(figsize = (12,8))
sns.heatmap(tab, annot = True)
plt.xlabel("Has Sidewalk Riding")
plt.ylabel("Type of Bike Lane")
plt.title("Effect of Bike Lane on Sidewalk Riding\n(in Percentage)")
plt.tight_layout()
plt.savefig(fig_path + 'bike_lane_and_sidewalk_riding', fmt = 'png', dpi = 300)
```

```python
temp = pd.DataFrame(df.groupby('location_id').sum()[['sidewalk', 'total_count']]).reset_index()
temp = pd.merge(temp, df.drop_duplicates('location_id')[['location_id', 'bikewaytype_options', 'city']])
temp['sidewalk_prop'] = temp['sidewalk'] / temp['total_count']
print(temp.head())

temp.groupby('bikewaytype_options').mean()['sidewalk_prop']
```
